```python
print("------------- Kobe Bryant ------------")
with open('Table1.csv') as csvfile:
    readCSV = csv.reader(csvfile, delimiter=',')
    for row in readCSV:
        t11_clt1.append(int(row[5]))
        t11_clt2.append(int(row[6]))
        t12_clt1.append(float(row[9])*100)
        t12_clt2.append(float(row[10])*100)
        t13_clt1.append(float(row[13])*100)
        t13_clt2.append(float(row[14])*100)
        
    std1=statistics.stdev(t11_clt1)
    std2=statistics.stdev(t11_clt2)
    std3=statistics.stdev(t12_clt1)
    std4=statistics.stdev(t12_clt2)
    std5=statistics.stdev(t13_clt1)
    std6=statistics.stdev(t13_clt2)
    print("Standard Deviation Kobe Bryant (All Field) CLTCH: ",std1, ", this is Clutch Shot taken." )
    print("----------------------------------")
    print("Standard Deviation Kobe Bryant (All Field) CLTCH2: ",std2, ", this is Clutch Shot made.")
    print("----------------------------------")
    print("Standard Deviation Kobe Bryant (Player Shots) CLTCH: ",std3, ", this is Clutch Shot % that had elevated impact on the game" )
    print("----------------------------------")
    print("Standard Deviation Kobe Bryant (Player Shots) CLTCH2: ",std4, ", this is Clutch Shot % that had crucial impact on the game (Buzzer Beaters.)")
    print("----------------------------------")
    print("Standard Deviation Kobe Bryant (Effective Field) CLTCH: ",std5, ", this is Effective field goals % that have elevated impact on the game ")
    print("----------------------------------")
    print("Standard Deviation Kobe Bryant (Effective Field) CLTCH2: ",std6, ", this is Effective field goals % that had crucial impact on the game (Buzzer Beaters.)")
print("----------------------------------")

print("------------- Lebron James ------------")


with open('Table2.csv') as csvfile:
    readCSV = csv.reader(csvfile, delimiter=',')
    for row in readCSV:
        t21_clt1.append(int(row[5]))
        t21_clt2.append(int(row[6]))
        t22_clt1.append(float(row[9])*100)
        t22_clt2.append(float(row[10])*100)
        t23_clt1.append(float(row[13])*100)
        t23_clt2.append(float(row[14])*100)

    std1=statistics.stdev(t21_clt1)
    std2=statistics.stdev(t21_clt2)
    std3=statistics.stdev(t22_clt1)
    std4=statistics.stdev(t22_clt2)
    std5=statistics.stdev(t23_clt1)
    std6=statistics.stdev(t23_clt2)
    print("Standard Deviation Lebron James (All Field) CLTCH: ",std1 , ", this is Clutch Shot taken." )
    print("----------------------------------")
    print("Standard Deviation Lebron James (All Field) CLTCH2: ",std2, ", this is Clutch Shot made.")
    print("----------------------------------")
    print("Standard Deviation Lebron James (Player Shots) CLTCH: ",std3, ", this is Clutch Shot % that had elevated impact on the game" )
    print("----------------------------------")
    print("Standard Deviation Lebron James (Player Shots) CLTCH2: ",std4, ", this is Clutch Shot % that had crucial impact on the game (Buzzer Beaters.)")
    print("----------------------------------")
    print("Standard Deviation Lebron James (Effective Field) CLTCH: ",std5, ", this is Effective field goals % that have elevated impact on the game ")
    print("----------------------------------")
    print("Standard Deviation Lebron James (Effective Field) CLTCH2: ",std6, ", this is Effective field goals % that had crucial impact on the game (Buzzer Beaters.)")
print("----------------------------------")
```

```python
def best_fit_slope(xs,ys):
    m = (((mean(xs)*mean(ys)) - mean(xs*ys)) /
         ((mean(xs)*mean(xs)) - mean(xs*xs)))
    return m

def best_fit_slope_and_intercept(xs,ys):
  a=mean(xs)
  b=mean(ys)
  temps=[]
  temps2=[]
  i=0
  while i < len(xs):
    temps.append(xs[i]*ys[i])
    temps2.append(xs[i]*xs[i])
    i+=1
  c=mean(temps)
  txs=[int(x) for x in xs]
  tys=[int(y) for y in ys]
  print(tys)
  m = ((a*b) - c)/((a*a) - mean(temps2))
  b = mean(tys) - m*mean(txs)

  return m, b
```

For this part of analysis, I try to plot a linear correlation of each player's clutch performances throughout the years, and try to visualize who is more consistent over the years. However, one big problem that I ran into is that when I read the CSV table file, I stored them in 1-D array, and it's too late to change all the arrays to 2-D arrays at this stage of development. So I found an alternative way to by writing functions that can help me calculate the slope and interception of my graph and I can use the slope and interception to graph my linear correlation fit for each data. 

```python
pyplot.title("Kobe(All Clutch Field Goals Attempted)")
pyplot.ylabel("Clutch 1")
pyplot.xlabel("Years")
pyplot.scatter(years,t11_clt1)
m, b = best_fit_slope_and_intercept(years,t11_clt1)
regression_line = [(m*x)+b for x in years]
pyplot.plot(years,regression_line)
pyplot.show()
```

First we see that Kobe has a negative linear correlation on all clutch field goals attempted over the years. As I mentioned in the previous content, Kobe was not the go-to guy during clutch time at the beginning of his career. He tended to pass the ball to Shaq during the first 5 years of his career. After Shaq left, he started to take over the game during clutch time. However, after 2010, he started to experiencing injuries and have to sit out a lot of times. Also, Lakers(Kobe's team) started to tank for higher draft pick. So a lot of the time, the game would just enter trash time instead of clutch time. That's why Kobe has a negative linear correlation relationship of his all field goals attempted.
